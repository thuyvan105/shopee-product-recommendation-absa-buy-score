```python
def find_csv_by_name(filename, search_root="/kaggle/input"):
    """Tìm file CSV theo tên trong thư mục Kaggle input."""
    for root, dirs, files in os.walk(search_root):
        if filename in files:
            return os.path.join(root, filename)
    return None


# Nếu chạy trên Kaggle
products_path = find_csv_by_name("shopee_products_clean.csv")
reviews_path = find_csv_by_name("shopee_reviews_clean.csv")

if products_path is None:
    products_path = "shopee_products_clean.csv"

if reviews_path is None:
    reviews_path = "shopee_reviews_clean.csv"

print("Products path:", products_path)
print("Reviews path:", reviews_path)
```

```python
products = pd.read_csv(products_path)
reviews = pd.read_csv(reviews_path)

print("Products shape:", products.shape)
print("Reviews shape:", reviews.shape)

display(products.head())
display(reviews.head())
```

```python
def normalize_unicode(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = unicodedata.normalize("NFC", text)
    return text


def normalize_name_for_matching(text):
    text = normalize_unicode(text)

    # Chuẩn hóa ký hiệu
    text = text.replace("™", "").replace("®", "")
    text = text.replace("gen", " gen ")

    # Chuẩn hóa ký tự phân tách
    text = re.sub(r"[\|\-_/,:;–—]+", " ", text)
    text = re.sub(r"[()\[\]{}]", " ", text)

    # Chuẩn hóa ký hiệu inch
    text = re.sub(r"[”\"']", " inch ", text)

    # Xóa từ nhiễu
    for phrase in sorted(NOISE_PHRASES, key=len, reverse=True):
        pattern = r"\b" + re.escape(phrase) + r"\b"
        text = re.sub(pattern, " ", text, flags=re.IGNORECASE)

    # Chuẩn hóa RAM/ROM thường gặp
    text = re.sub(r"(\d+)\s*\+\s*(\d+)\s*gb", r"\1gb \2gb", text)
    text = re.sub(r"(\d+)\s*/\s*(\d+)\s*gb", r"\1gb \2gb", text)
    text = re.sub(r"(\d+)\s*gb\s*/\s*(\d+)\s*gb", r"\1gb \2gb", text)

    # Chuẩn hóa khoảng trắng
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
def extract_storage_options(text):
    text = normalize_unicode(text)

    # Chuẩn hóa dạng 8+256GB, 12GB/256GB
    text = re.sub(r"(\d+)\s*\+\s*(\d+)\s*gb", r"\1gb \2gb", text)
    text = re.sub(r"(\d+)\s*/\s*(\d+)\s*gb", r"\1gb \2gb", text)

    matches = re.findall(r"(\d+(?:\.\d+)?)\s*(tb|gb)", text)

    storages = []

    for num, unit in matches:
        val = float(num)

        if unit == "tb":
            if val.is_integer():
                storages.append(f"{int(val)}tb")
            else:
                storages.append(f"{val}tb")

        else:
            # Lọc RAM nhỏ như 4GB, 6GB, 8GB, 12GB, 16GB.
            # Chỉ giữ các giá trị thường là ROM/storage: >= 32GB.
            if val >= 32:
                storages.append(f"{int(val)}gb")

    # Unique nhưng giữ thứ tự xuất hiện
    unique_storages = []
    for s in storages:
        if s not in unique_storages:
            unique_storages.append(s)

    return unique_storages
```

```python
def extract_product_type(name):
    text = normalize_unicode(name)

    # Đặt phụ kiện lên trước để tránh trường hợp "Cáp sạc Apple Watch" bị gom nhầm vào đồng hồ.
    rules = [
        ("charger_cable_accessory", [
            "cáp", "sạc", "adapter", "keyboard", "pencil", "pen",
            "ốp", "case", "phụ kiện", "power adapter", "usb"
        ]),
        ("phone", [
            "điện thoại", "iphone", "galaxy a", "galaxy s",
            "oppo", "vivo", "tecno spark", "poco", "redmi note"
        ]),
        ("tablet", [
            "máy tính bảng", "ipad", "matepad", "galaxy tab",
            "redmi pad", "xiaomi pad", "poco pad"
        ]),
        ("smartwatch_band", [
            "đồng hồ", "watch", "band", "vòng đeo tay"
        ]),
        ("earphone_speaker", [
            "tai nghe", "buds", "freebuds", "airpods", "sony wf", "sony wh", "jbl", "loa"
        ]),
        ("home_appliance", [
            "máy giặt", "máy hút bụi", "nồi", "máy xay", "quạt", "bear"
        ]),
    ]

    for label, keywords in rules:
        if any(keyword in text for keyword in keywords):
            return label

    return "other"
```

```python
def first_match(patterns, text):
    for pattern in patterns:
        match = re.search(pattern, text, flags=re.IGNORECASE)
        if match:
            value = match.group(1)
            value = re.sub(r"\s+", " ", value).strip()
            return value
    return None


def clean_base_model(base):
    if not base:
        return ""

    base = normalize_name_for_matching(base)

    # Chuẩn hóa một số trường hợp Apple
    base = base.replace("4th gen", "4")
    base = base.replace("airpods 4th", "airpods 4")

    # Cắt bớt mô tả kỹ thuật bị kéo theo
    base = re.sub(
        r"\b(?:màn|màn hình|pin|camera|thiết kế|màu|inch|wifi|cellular|chống|kháng|thời lượng|nghe gọi|loa kép)\b.*$",
        "",
        base
    )

    base = re.sub(r"\s+", " ", base).strip()
    return base


def extract_base_model(name, brand=None):
    # Dùng cả text gốc và text đã normalize.
    # Một số từ như "cáp sạc" bị xóa trong normalize vì là từ nhiễu,
    # nên phải xử lý phụ kiện trước bằng text gốc.
    raw_text = normalize_unicode(name)
    text = normalize_name_for_matching(name)

    # Accessory patterns first, to avoid grouping cables/chargers as Apple Watch/iPad/etc.
    if ("apple watch" in raw_text) and any(k in raw_text for k in ["cáp", "sạc", "charger", "usb"]):
        return "cáp sạc apple watch usb c"

    if ("apple" in raw_text) and ("20w" in raw_text) and ("adapter" in raw_text or "sạc" in raw_text):
        return "apple 20w usb c power adapter"

    if ("lightning" in raw_text) and ("cáp" in raw_text or "cable" in raw_text):
        return "apple usb c to lightning cable"

    accessory_patterns = [
        r"\b(magic\s*keyboard\s*\d*)\b",
        r"\b(poco\s*smart\s*pen)\b",
        r"\b(poco\s*pad\s*smart\s*keyboard)\b",
        r"\b(sạc\s*xiaomi\s*\d+w\s*hypercharge)\b",
    ]

    acc = first_match(accessory_patterns, text)
    if acc:
        return clean_base_model(acc)

    patterns = [
        # Apple
        r"\b(iphone\s*(?:air|se|\d{1,2})\s*(?:pro max|pro|plus|max)?)\b",
        r"\b(apple\s*iphone\s*(?:air|se|\d{1,2})\s*(?:pro max|pro|plus|max)?)\b",
        r"\b(ipad\s*(?:air|pro|mini|a\d+)?\s*(?:\d{4})?)\b",
        r"\b(apple\s*watch\s*series\s*\d+\s*(?:gps)?\s*(?:\d+mm)?)\b",
        r"\b(apple\s*watch\s*(?:se\s*)?\d*(?:\s*gps)?(?:\s*\d+mm)?)\b",
        r"\b(airpods\s*(?:pro|max)?\s*(?:\d+(?:th)?(?:\s*gen)?)?)\b",

        # Samsung
        r"\b(samsung\s*galaxy\s*(?:s\d+|a\d+|m\d+|z\s*flip\d*|z\s*fold\d*|tab\s*[a-z0-9]+)\s*(?:ultra|plus|fe|lte|5g|4g)?)\b",
        r"\b(galaxy\s*(?:s\d+|a\d+|m\d+|z\s*flip\d*|z\s*fold\d*|tab\s*[a-z0-9]+)\s*(?:ultra|plus|fe|lte|5g|4g)?)\b",

        # Xiaomi / Redmi / POCO
        r"\b(poco\s*pad\s*(?:[a-z0-9]+)?)\b",
        r"\b(poco\s*[a-z]?\d+\s*(?:pro|ultra|lite)?\s*(?:5g)?)\b",
        r"\b(redmi\s*note\s*\d+\s*(?:pro max|pro|plus)?\s*(?:5g)?)\b",
        r"\b(redmi\s*a\d+\s*(?:pro)?\s*(?:5g)?)\b",
        r"\b(redmi\s*pad\s*\d*(?:\s*wi fi)?)\b",
        r"\b(redmi\s*buds\s*\d+\s*(?:play|pro|active)?)\b",
        r"\b(redmi\s*watch\s*\d+\s*(?:active|lite)?)\b",
        r"\b(xiaomi\s*(?:smart\s*band|watch|pad)?\s*\d+\s*(?:active|pro|ultra|lite)?\s*(?:5g)?)\b",
        r"\b(xiaomi\s*\d+\s*(?:pro|ultra|lite)?\s*(?:5g)?)\b",

        # OPPO
        r"\b(oppo\s*(?:reno\s*\d+|a\d+|find\s*[a-z0-9]+|pad|enco\s*[a-z0-9 ]*)\s*(?:pro|plus|flip|5g|4g)?)\b",

        # vivo
        r"\b(vivo\s*(?:v\d+|y\d+)\s*(?:pro|fe|5g|4g)?)\b",

        # TECNO
        r"\b(tecno\s*(?:spark\s*go\s*\d+|spark\s*\d+|pova\s*\d+|camon\s*\d+|buds\s*\d+)\s*(?:pro|go|5g)?)\b",

        # HUAWEI
        r"\b(huawei\s*matepad\s*[a-z0-9. ]{0,20})\b",
        r"\b(huawei\s*watch\s*(?:fit\s*)?\d*(?:\s*series|\s*d2)?)\b",
        r"\b(huawei\s*freebuds\s*[a-z0-9 ]{0,15})\b",
        r"\b(huawei\s*band\s*\d+)\b",
        r"\b(huawei\s*(?:freeclip|freearc))\b",

        # realme
        r"\b(realme\s*(?:watch|c|note|buds)?\s*[a-z0-9 ]{0,15})\b",

        # Audio / accessories / home appliances
        r"\b(sony\s*(?:wf|wh)[a-z0-9\-]+)\b",
        r"\b(jbl\s*[a-z0-9 ]{0,20})\b",
        r"\b(belkin\s*[a-z0-9 ]{0,20})\b",
        r"\b(bear\s*[a-z0-9\- ]{0,25})\b",
    ]

    base = first_match(patterns, text)

    if base:
        return clean_base_model(base)

    # Fallback nếu không bắt được model rõ ràng
    tokens = re.findall(r"[a-z0-9à-ỹ]+", text)

    if brand is not None and not pd.isna(brand):
        brand_clean = normalize_unicode(brand)
        if brand_clean and brand_clean not in tokens:
            tokens = [brand_clean] + tokens

    return " ".join(tokens[:6]) if tokens else text
```

```python
def build_canonical_name(row):
    name = row.get("product_name", "")
    brand = row.get("brand", "")

    base = extract_base_model(name, brand)
    storage_options = extract_storage_options(name)

    product_type = row.get("product_type", "")

    # Chỉ thêm dung lượng cho phone/tablet. Phụ kiện như cáp/sạc không cần thêm storage.
    if product_type in ["phone", "tablet"] and len(storage_options) == 1:
        canonical = f"{base} {storage_options[0]}"
    else:
        canonical = base

    canonical = normalize_name_for_matching(canonical)
    canonical = re.sub(r"\s+", " ", canonical).strip()

    return canonical


products["product_type"] = products["product_name"].apply(extract_product_type)
products["product_name_norm"] = products["product_name"].apply(normalize_name_for_matching)
products["storage_options"] = products["product_name"].apply(lambda x: ", ".join(extract_storage_options(x)))
products["canonical_product_name"] = products.apply(build_canonical_name, axis=1)

# Tạo ID nhóm sản phẩm
products["canonical_product_id"] = products["canonical_product_name"].astype("category").cat.codes

display(products[
    [
        "shop_clean", "brand", "product_type",
        "product_name", "product_name_norm",
        "storage_options", "canonical_product_name", "canonical_product_id"
    ]
].head(30))
```

```python
canonical_group_summary = (
    products
    .groupby(["canonical_product_id", "canonical_product_name"])
    .agg(
        num_items=("item_id", "nunique"),
        num_shops=("shop_clean", "nunique"),
        product_type=("product_type", lambda x: x.mode().iloc[0] if len(x.mode()) else "unknown"),
        min_price=("current_price", "min"),
        max_price=("current_price", "max"),
        avg_rating=("rating", "mean"),
        total_sold=("sold_count_num", "sum"),
        example_names=("product_name", lambda x: " || ".join(x.head(5)))
    )
    .reset_index()
    .sort_values(["num_shops", "num_items", "total_sold"], ascending=False)
)

display(canonical_group_summary.head(30))
```

```python
same_product_candidates = canonical_group_summary[
    canonical_group_summary["num_shops"] >= 2
].copy()

print("Số nhóm sản phẩm có từ 2 shop trở lên:", same_product_candidates.shape[0])
display(same_product_candidates.head(30))
```

```python
import os
import pandas as pd
import numpy as np
import torch
from transformers import pipeline
from tqdm.auto import tqdm

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_colwidth", 120)
```

```python
PRODUCTS_PATH = "/kaggle/working/shopee_products_processed.csv"
REVIEWS_PATH = "/kaggle/working/shopee_reviews_processed.csv"

products = pd.read_csv(PRODUCTS_PATH)
reviews = pd.read_csv(REVIEWS_PATH)

print("Products:", products.shape)
print("Reviews:", reviews.shape)

display(products.head())
display(reviews.head())
```

```python
ASPECT_MODELS = {
    "BATTERY": "van105/absa-visfd-final-battery",    
    "CAMERA": "van105/absa-visfd-final-camera",
    "SCREEN": "van105/absa-visfd-final-screen",
    "PERFORMANCE": "van105/absa-visfd-final-performance",
    "PRICE": "van105/absa-visfd-final-price",
    "DESIGN": "van105/absa-visfd-final-design",
    "FEATURES": "van105/absa-visfd-final-features",
    "STORAGE": "van105/absa-visfd-final-storage",
    "SERANDACC": "van105/absa-visfd-final-serandacc",
    "GENERAL": "van105/absa-visfd-final-general",
}

ASPECTS = list(ASPECT_MODELS.keys())

LABEL_MAP = {
    "LABEL_0": "negative",
    "LABEL_1": "neutral",
    "LABEL_2": "positive",
    "LABEL_3": "not_mentioned"
}

device = 0 if torch.cuda.is_available() else -1
print("Device:", "GPU" if device == 0 else "CPU")
```

```python
TEXT_COL = "review_text_clean"

if TEXT_COL not in reviews.columns:
    raise ValueError(f"Không tìm thấy cột {TEXT_COL}. Hãy kiểm tra file shopee_reviews_processed.csv")

reviews[TEXT_COL] = reviews[TEXT_COL].fillna("").astype(str)
reviews = reviews[reviews[TEXT_COL].str.strip() != ""].copy()

# Test nhanh thì đặt SAMPLE_N = 500 hoặc 1000.
# Chạy thật thì để SAMPLE_N = None.
SAMPLE_N = None

if SAMPLE_N is not None:
    reviews_run = reviews.sample(n=min(SAMPLE_N, len(reviews)), random_state=42).copy()
else:
    reviews_run = reviews.copy()

texts = reviews_run[TEXT_COL].tolist()

print("Số review chạy ABSA:", len(texts))
display(reviews_run.head())
```

```python
def predict_aspect_labels(texts, model_name, batch_size=32):
    clf = pipeline(
        "text-classification",
        model=model_name,
        tokenizer=model_name,
        device=device,
        truncation=True,
        max_length=256
    )

    results = clf(
        texts,
        batch_size=batch_size,
        truncation=True,
        max_length=256
    )

    labels = [LABEL_MAP.get(r["label"], r["label"]) for r in results]
    scores = [float(r["score"]) for r in results]
    return labels, scores
```

```python
for aspect, model_name in ASPECT_MODELS.items():
    print(f"Running aspect model: {aspect} - {model_name}")

    labels, scores = predict_aspect_labels(
        texts=texts,
        model_name=model_name,
        batch_size=32
    )

    reviews_run[f"{aspect}_label"] = labels
    reviews_run[f"{aspect}_score"] = scores

print("ABSA labeling done!")
display(reviews_run.head())
```

```python
reviews_run.to_csv("/kaggle/working/shopee_reviews_absa_labeled.csv", index=False, encoding="utf-8-sig")
print("Saved: /kaggle/working/shopee_reviews_absa_labeled.csv")
```

```python
base_cols = [
    "shop",
    "shop_id",
    "item_id",
    "product_name",
    "review_star",
    TEXT_COL
]

# Giữ thêm canonical nếu đã có trong reviews_processed
for col in ["canonical_product_name", "canonical_product_id", "product_type"]:
    if col in reviews_run.columns and col not in base_cols:
        base_cols.append(col)

long_rows = []

for aspect in ASPECTS:
    temp = reviews_run[base_cols + [f"{aspect}_label", f"{aspect}_score"]].copy()
    temp["aspect"] = aspect
    temp["sentiment"] = temp[f"{aspect}_label"]
    temp["confidence"] = temp[f"{aspect}_score"]
    temp = temp.drop(columns=[f"{aspect}_label", f"{aspect}_score"])
    long_rows.append(temp)

review_aspect_long = pd.concat(long_rows, ignore_index=True)

review_aspect_mentioned = review_aspect_long[
    review_aspect_long["sentiment"] != "not_mentioned"
].copy()

print("review_aspect_long:", review_aspect_long.shape)
print("review_aspect_mentioned:", review_aspect_mentioned.shape)

display(review_aspect_mentioned.head())
```

```python
sentiment_summary = (
    review_aspect_mentioned
    .groupby(["shop_id", "item_id", "aspect", "sentiment"])
    .size()
    .reset_index(name="count")
)

sentiment_pivot = sentiment_summary.pivot_table(
    index=["shop_id", "item_id", "aspect"],
    columns="sentiment",
    values="count",
    fill_value=0
).reset_index()

for col in ["negative", "neutral", "positive"]:
    if col not in sentiment_pivot.columns:
        sentiment_pivot[col] = 0

sentiment_pivot["total_mentions"] = (
    sentiment_pivot["negative"] +
    sentiment_pivot["neutral"] +
    sentiment_pivot["positive"]
)

sentiment_pivot = sentiment_pivot[sentiment_pivot["total_mentions"] > 0].copy()

sentiment_pivot["positive_rate"] = sentiment_pivot["positive"] / sentiment_pivot["total_mentions"]
sentiment_pivot["negative_rate"] = sentiment_pivot["negative"] / sentiment_pivot["total_mentions"]
sentiment_pivot["neutral_rate"] = sentiment_pivot["neutral"] / sentiment_pivot["total_mentions"]

sentiment_pivot["aspect_sentiment_score"] = (
    sentiment_pivot["positive"] - sentiment_pivot["negative"]
) / sentiment_pivot["total_mentions"]

product_info_cols = [
    "shop", "shop_id", "item_id", "product_name",
    "canonical_product_name", "canonical_product_id",
    "product_type", "brand", "current_price", "rating",
    "sold_count_num", "monthly_sold_num", "product_url"
]
product_info_cols = [c for c in product_info_cols if c in products.columns]

product_aspect_summary = sentiment_pivot.merge(
    products[product_info_cols].drop_duplicates(["shop_id", "item_id"]),
    on=["shop_id", "item_id"],
    how="left"
)

display(product_aspect_summary.head())
```

```python
ASPECT_WEIGHTS = {
    "GENERAL": 1.0,
    "PRICE": 1.3,
    "PERFORMANCE": 1.3,
    "BATTERY": 1.2,
    "SCREEN": 1.1,
    "CAMERA": 1.0,
    "DESIGN": 0.9,
    "FEATURES": 1.0,
    "STORAGE": 0.8,
    "SERANDACC": 1.1
}

sentiment_pivot["aspect_weight"] = sentiment_pivot["aspect"].map(ASPECT_WEIGHTS).fillna(1.0)
sentiment_pivot["weighted_score"] = (
    sentiment_pivot["aspect_sentiment_score"] *
    sentiment_pivot["aspect_weight"] *
    sentiment_pivot["total_mentions"]
)

product_absa_score = (
    sentiment_pivot
    .groupby(["shop_id", "item_id"])
    .agg(
        absa_weighted_sum=("weighted_score", "sum"),
        absa_weight_sum=("total_mentions", "sum"),
        total_aspect_mentions=("total_mentions", "sum"),
        avg_positive_rate=("positive_rate", "mean"),
        avg_negative_rate=("negative_rate", "mean"),
        avg_neutral_rate=("neutral_rate", "mean")
    )
    .reset_index()
)

product_absa_score["absa_score_raw"] = (
    product_absa_score["absa_weighted_sum"] / product_absa_score["absa_weight_sum"]
)

product_absa_score["absa_score"] = (
    (product_absa_score["absa_score_raw"] + 1) / 2
) * 100

display(product_absa_score.head())
```
